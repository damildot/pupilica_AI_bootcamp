```python
import pandas as pd
veri = pd.read_csv("breast-cancer-data.csv")
veri
```

```python
print(veri.head()) #ilk beş satır
print("-------------------------------")
print(veri.tail()) #son 5 satır
print("-------------------------------")
print(veri.info()) #dataframe ile ilgili temel bilgilere erişmemiz lazım
print("-------------------------------")
print(veri.dtypes) # display the data types of each column in the Pandas DataFrame
print("-------------------------------")
print(veri.columns) #verinin sütunları
print("-------------------------------")
print(veri.axes)
print("-------------------------------")
print(veri.describe()) # veri seti icerisinde bulunan numeric feature'larin temel istatistiksel analizi
```

```python
#line plot
import matplotlib.pyplot as plt
plt.plot(veri.texture_mean, color = "red", alpha = 0.5, label  = "texture_mean")
plt.xlabel("x etiketi")
plt.ylabel("y etiketi")
plt.title("texture_mean")
plt.legend()
plt.show()
```

```python
#scatter plot
plt.scatter(veri.radius_mean, veri.area_mean, color = "blue", s = 15, alpha = 0.5)
plt.xlabel("radius_mean")
plt.ylabel("area_mean")
plt.title("baslik")
```
