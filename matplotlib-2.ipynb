```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(veri["SepalLengthCm"], color = "red", alpha = 0.9, bins=20 , label = "SepalLengthCm")
plt.hist(veri["PetalLengthCm"], color = "blue", alpha = 0.9, bins=20 , label = "PetalLengthCm")

plt.ylabel("Frekans (KaÃ§ adet olduÄŸu)")
plt.xlabel("cm")
plt.legend()
```

```python
import pandas as pd
veri = pd.read_csv("iris.csv")
veri
```

```python
fig, ax = plt.subplots(2,1) # 2 satir, 1 sutun
fig.suptitle("Ana baslik")
ax[0].plot(veri.SepalLengthCm, color = "red") # ax'lerin sifirincisi ilk eksen birinci plot
ax[0].set_title("Sepal Length")
ax[0].set_xlabel("index")
ax[0].set_ylabel("cm")

ax[1].plot(veri.PetalLengthCm, color = "blue")
ax[1].set_title("Petal Length")
ax[1].set_xlabel("index")
ax[1].set_ylabel("cm")

fig.tight_layout() # matplotlib kÃ¼tÃ¼phanesinde alt grafikler arasÄ±ndaki boÅŸluklarÄ± otomatik olarak ayarlar.
fig.subplots_adjust(top = 0.85)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 2x2'lik bir grid iÃ§inde 4 alt grafik oluÅŸtur
plt.subplot(2, 2, 1) # 2 satir ve 2 sutundan olussun ve ben 1. plotu cizdiriyorum
plt.scatter([1, 2, 3, 4], [5, 6, 7, 8])
plt.title("Alt Grafik 1")

plt.subplot(2, 2, 2)
plt.plot([1, 2, 3, 4], [8, 7, 6, 5])
plt.title("Alt Grafik 2")

plt.subplot(2, 2, 3) # 2 satir ve 2 sutundan olussun ve ben 3. plotu cizdiriyorum
plt.hist([1, 2, 3, 4, 5, 6, 7, 8])
plt.title("Alt Grafik 3")

plt.subplot(2, 2, 4)
plt.plot([1, 2, 3, 4], [8, 7, 6, 5])
plt.title("Alt Grafik 4")

plt.show() # Grafikleri gÃ¶ster
```

```python
# dataseti indir: https://github.com/prashant-kikani/breast-cancer-detection/blob/master/breast-cancer-data.csv
# class'larin (target variable) diagnosis ayrilmasi
# 2,3 bir subplot cizdirelim
# ilk satirda     M iÃ§in scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
# ikinici satirda B iÃ§in scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
```

```python
import pandas as pd
data = pd.read_csv("breast-cancer-data.csv")
data
```

```python
# class'larin (target variable) diagnosis ayrilmasi
m = data[data["diagnosis"] == "M"]
m.reset_index(drop=True, inplace=True)
b = data[data["diagnosis"] == "B"]
b.reset_index(drop=True, inplace=True)

# ilk satirda     M iÃ§in scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
plt.subplot(2, 3, 1)
plt.scatter(m.radius_mean, m.texture_mean)

plt.subplot(2, 3, 2)
plt.hist(m.radius_mean)

plt.subplot(2, 3, 3)
plt.plot(m.radius_mean)

# ikinici satirda B iÃ§in scatter(radius_mean vs texture_mean), hist(radius_mean), plot(radius_mean)
plt.subplot(2, 3, 4)
plt.scatter(b.radius_mean, b.texture_mean)

plt.subplot(2, 3, 5)
plt.hist(b.radius_mean)

plt.subplot(2, 3, 6)
plt.plot(b.radius_mean)
```
